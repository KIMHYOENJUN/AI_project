```python
import SimpleITK as sitk
from skimage.transform import rotate
import cv2 as cv
from keras.preprocessing import image as kimg
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
from matplotlib.ticker import MultipleLocator
```

```python
test_data = './Deepnoid-Education/dataset/test.dcm'
#############################
# fill the blank
# original image 와 augmentation 한 image를 4장을 만드세요 
# 5장의 이미지를 하나의 array로 만드세요
# array를 마지막에 있는 display의 argument로 넣어서 현재 출력된 이미지와 유사하도록 만드세요
```

```python
# display 함수는 수정하지 말고 그대로 사용해주세요.
def display(array):
  import matplotlib.pyplot as plt
  for i in range(array.shape[0]):
    plt.imshow(array[i], cmap='gray')
    plt.show()
```

```python
from IPython.core.pylabtools import figsize

image = sitk.ReadImage(test_data)
img_arr = sitk.GetArrayFromImage(image)

plt.figure(figsize=(4,4))
plt.imshow(img_arr[0], cmap='gray')

img_rot = rotate(img_arr[0],20)
plt.figure(figsize=(4,4))
plt.imshow(img_rot, cmap='gray')
plt.show()

img_shift = kimg.random_shift(np.array(img_arr),0.3, 0.3)
plt.figure(figsize=(4,4))
plt.imshow(img_shift[0], cmap='gray')
plt.show()

img_hflip = np.fliplr(img_arr[0])
plt.figure(figsize=(4,4))
plt.imshow(img_hflip, cmap='gray')
plt.show()

#img_vflip = np.flipud(img_arr[0])
#plt.figure(figsize=(4,4))
#plt.imshow(img_vflip, cmap='gray')
#plt.show()
```
